```python
from sklearn.datasets import make_blobs


X, y = make_blobs(n_samples=150, 
                  n_features=2, 
                  centers=4, 
                  cluster_std=0.5, 
                  shuffle=True, 
                  random_state=0)
```

```python
def similiariy(x: list, y: list) -> int:
    """compute similiarity between two points"""

    square_diff = [(i - j)**2 for i, j in zip(x, y)]
    print(f'D² or Similarity => {sum(square_diff)}')
    

i = [1, 2, 3]
j = [10, 23, 21]

similiariy(i, j)
```

```python
import numpy as np
from typing import List, Tuple

def sse(points: List[Tuple[float, float]], centroid: Tuple[float, float]) -> float:
    """
    Calcula a soma dos erros quadráticos (SSE) para um grupo de pontos em relação ao centróide.
    """
    total = 0.0 # bagunça
    for p in points:
        dist_squared = np.linalg.norm(np.array(p) - np.array(centroid)) ** 2
        total += dist_squared
        print(f'Ponto: {p}, Distância² para centróide {centroid}: {dist_squared:.2f}')
    
    print(f'\nSSE total: {total:.2f}')
    return total

# Exemplo de uso
centroide = [(0, 0), (5,4)]
grupo = [(1.0, 2.0), (2.0, 3.0), (0.0, 1.0)]

sse(grupo, centroide)
```

```python
import matplotlib.pyplot as plt

# Example data
points = centroide
grupos = grupo

# Separate x and y coordinates
x_coords, y_coords = zip(*points)
x_g, y_g = zip(*grupos)

# Plot as points
plt.scatter(x_coords, y_coords)
plt.scatter(x_g, y_g)

# Show the plot
plt.show()
```
